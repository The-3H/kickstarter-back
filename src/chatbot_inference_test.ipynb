```python
from chatbot import ChatQA

chat_qa = ChatQA()


query = """
Lately, I've been feeling like I'm getting gas whenever I eat, 
and it's like my stomach is full and tight. What can I do?
"""
input = {
    'age' : "40",
    'pregnancy_week': "16",
    'input': query
}

response = chat_qa.get_result(input['age'], input['pregnancy_week'], input['input'])
```
